## **Positional Embedding trong Transformer**

```python
import math
import numpy as np

def create_positional_matrix(seq_len, embed_size):
    position_matrix = [[0 for _ in range(embed_size)] for _ in range(seq_len)]
    for pos in range(seq_len):
        for i in range(embed_size):
            if i % 2 == 0:
                position_matrix[pos][i] = math.sin(pos / (10000 ** (i/embed_size)))
            else:
                position_matrix[pos][i] = math.cos(pos / (10000 ** ((i-1)/embed_size)))
    return position_matrix


seq_length = 10
embed_size = 16
position_matrix = create_positional_matrix(seq_length, embed_size)
print(f"Position matrix: \n {np.array(position_matrix)}")
```
