# 1. Sigmoid function:

```python
import math

def sigmoid(list_data):
    result = []
    for value in list_data:
        result.append(round(1 / (1 + math.exp(-value)), 3))
    return result

data = [1, 5, -4, 3, -2]
sigmoid(data)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(sigmoid(data))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 2. Tanh Function:

```python
import math

def tanh(list_data):
    result = []
    for value in list_data:
        result.append(round((2 / (1 + math.exp(-2 * value))) - 1, 3))
    return result

data = [1, 5, -4, 3, -2]
tanh(data)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(tanh(data))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 3. Softplus function:

```python
import math

def softplus(list_data):
    result = []
    for value in list_data:
        result.append(round(math.log(1 + math.exp(value)), 3))
    return result

data = [1, 5, -4, 3, -2]
softplus(data)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(softplus(data))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('Softplus Function')
plt.xlabel('x')
plt.ylabel('softplus(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 4. ReLU Function:

```python
def ReLU(list_data):
    result = []
    for value in list_data:
        result.append(max(0, value))
    return result

data = [1, 5, -4, 3, -2]
ReLU(data)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(ReLU(data))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 5. LeakyReLU Function:

```python
def LeakyReLU(list_data):
    result = []
    for value in list_data:
        result.append(max(0.01 * value, value))
    return result

data = [1, 5, -4, 3, -2]
LeakyReLU(data)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(LeakyReLU(data))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('Leaky Function')
plt.xlabel('x')
plt.ylabel('leaky(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 6. ELU Function:

```python
def ELU(list_data, alpha):
    result = []
    for value in list_data:
        result.append(round(max(alpha * (math.exp(value) - 1), value), 3))
    return result

data = [1, 5, -4, 3, -2]
ELU(data, 0.1)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(ELU(data, 0.1))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('ELU Function')
plt.xlabel('x')
plt.ylabel('ELU(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 7. PReLU Function:

```python
def PReLU(list_data, alpha):
    result = []
    for value in list_data:
        result.append(round(max(alpha * value, value), 3))
    return result

data = [1, 5, -4, 3, -2]
PReLU(data, 0.1)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(PReLU(data, 0.1))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('PReLU Function')
plt.xlabel('x')
plt.ylabel('PReLU(x)')
plt.grid(True)

# Show the plot
plt.show()
```

# 8. Swish function:

```python
def swish(list_data, alpha):
    result = []
    for value in list_data:
        result.append(round(value / (1 + math.exp(-alpha * value)), 3))
    return result

data = [1, 5, -4, 3, -2]
swish(data, 1)
```

```python
import matplotlib.pyplot as plt

x_values = sorted(data)

y_values = sorted(swish(data, 1))

# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Customize the plot (optional)
plt.title('Swish Function')
plt.xlabel('x')
plt.ylabel('swish(x)')
plt.grid(True)

# Show the plot
plt.show()
```
