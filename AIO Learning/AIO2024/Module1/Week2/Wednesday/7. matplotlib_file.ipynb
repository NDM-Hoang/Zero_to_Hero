```python
# ví dụ 2
import matplotlib.pyplot as plt
%matplotlib qt

# hàm linear_space
# num should be greater than 1
def linear_space(start, stop, num=10):    
    step = (stop - start) / num        

    result = []
    for i in range(num):
        result.append(start + step*i)        
    return result
```

```python
x_data = linear_space(-8, 8, 100)
print(x_data)
```

```python
import matplotlib.pyplot as plt

x_data = linear_space(-8, 8, 100)
y_data = [v*v for v in x_data]

plt.plot(x_data, y_data)
plt.ylabel('y values')
plt.xlabel('x values')

# hiển thị biểu đồ
plt.show()
```
